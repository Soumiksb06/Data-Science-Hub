##### Exercise 5. Write a Python program which accepts a string containing alphabets and digits and then displays the string in which the digits only reversed.

#### For example if the string is "Abcd1234" then the output should be "Abcd4321" and if the string is "Ab23cd56" then the output should be "Ab32cd65".

```python
# Input from the user
seq=''
result=''
input_str = input("Enter an alphanumeric string: ")
for char in input_str:
        if char.isdigit():
            seq+=char
        else:
            result+=seq[::-1]
            result+=char
            seq=''
result+=seq[::-1]
print("Resultant string with reversed numbers: ", result)
```
