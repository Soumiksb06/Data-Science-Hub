```python
import pandas as pd
import chardet
from collections import Counter
import re
import spacy
import json
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Load spaCy model for NLP tasks
nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])

# Function to detect file type and characteristics
def detect_file_characteristics(file_path):
    # Check if the file is CSV
    if file_path.endswith('.csv'):
        file_type = 'csv'

        # Detect encoding for CSV
        with open(file_path, 'rb') as myfile:
            raw_data = myfile.read()
            detected = chardet.detect(raw_data)
            encoding_type = detected['encoding']
            print(f"Detected encoding: {encoding_type}")

            try:
                df = pd.read_csv(file_path, encoding=encoding_type)
            except UnicodeDecodeError:
                print(f"Failed to decode with encoding {encoding_type}, trying 'ISO-8859-1'")
                encoding_type = 'ISO-8859-1'
                df = pd.read_csv(file_path, encoding=encoding_type)

        return df

    else:
        print("Unsupported file format. Please provide a CSV file.")
        return None

# Function to preprocess text
def preprocess_text(text):
    # Tokenization and cleaning using spaCy
    doc = nlp(text)
    tokens = [token.lemma_.lower() for token in doc if not token.is_stop and token.is_alpha]
    return tokens

# Function to process the data and extract topics using LDA
def extract_topics(df):
    # Concatenate Title 1 and Meta Description 1 into a single text column
    df['text'] = df['Title 1'].fillna('') + ' ' + df['Meta Description 1'].fillna('')

    # Preprocess text
    df['processed_text'] = df['text'].apply(preprocess_text)

    # Convert processed text into a format suitable for LDA
    text_data = df['processed_text'].apply(lambda x: ' '.join(x))

    # Use CountVectorizer to convert text into a matrix of token counts
    vectorizer = CountVectorizer(max_df=0.95, min_df=2, max_features=1000, stop_words='english')
    X = vectorizer.fit_transform(text_data)

    # Apply LDA
    lda = LatentDirichletAllocation(n_components=5, random_state=42)
    lda.fit(X)

    # Extract topics and associated words
    feature_names = vectorizer.get_feature_names_out()
    topics = {}
    for topic_idx, topic in enumerate(lda.components_):
        top_features_idx = topic.argsort()[:-10 - 1:-1]
        topics[f"Topic {topic_idx + 1}"] = [feature_names[i] for i in top_features_idx]

    return topics

# Example usage
input_file = '/content/drive/MyDrive/Competitors Data.xlsx'  # Replace with your file path
df = detect_file_characteristics(input_file)

if df is not None:
    # Preprocess the data to fill NaN values
    df = df.fillna('Not found')

    # Extract topics using LDA
    extracted_topics = extract_topics(df)

    # Convert the extracted topics to JSON format for easier reading
    topics_json = json.dumps(extracted_topics, indent=4)
    print(topics_json)

    # Save to JSON file
    output_json_path = '/content/extracted_topics.json'
    with open(output_json_path, 'w') as f:
        json.dump(extracted_topics, f, indent=4)
    print(f"Extracted topics saved to: {output_json_path}")
```

```python
import pandas as pd
import chardet
import csv
import json
import re
from collections import Counter
from sentence_transformers import SentenceTransformer

# Function to detect file type and characteristics
def detect_file_characteristics(file_path):
    if file_path.endswith('.csv'):
        file_type = 'csv'
        with open(file_path, 'rb') as myfile:
            raw_data = myfile.read()
            detected = chardet.detect(raw_data)
            encoding_type = detected['encoding']
            print(f"Detected encoding: {encoding_type}")

            try:
                df = pd.read_csv(file_path, encoding=encoding_type)
            except UnicodeDecodeError:
                print(f"Failed to decode with encoding {encoding_type}, trying 'ISO-8859-1'")
                encoding_type = 'ISO-8859-1'
                df = pd.read_csv(file_path, encoding=encoding_type)

        return df

    elif file_path.endswith('.xlsx') or file_path.endswith('.xls'):
        file_type = 'excel'
        df = pd.read_excel(file_path)
        return df

    else:
        print("Unsupported file format. Please provide a CSV or Excel file.")
        return None

# Function to embed text into vectors using Sentence Transformers
def embed_text(texts):
    model = SentenceTransformer('paraphrase-distilroberta-base-v1')
    embeddings = model.encode(texts, convert_to_tensor=True)
    return embeddings

# Function to process the data and identify topics dynamically
def process_data(df):
    titles = df['Title 1'].fillna('Not found').tolist()
    meta_descriptions = df['Meta Description 1'].fillna('Not found').tolist()

    # Combine titles and meta descriptions
    all_text = titles + meta_descriptions

    # Embed all text into vectors
    embeddings = embed_text(all_text)

    # Initialize a Counter for identified topics
    identified_topics = Counter()

    # Loop through embeddings to find similar topics
    for i in range(len(all_text)):
        topic_found = False
        for topic, count in identified_topics.items():
            if embeddings[i].dot(embeddings[count['index']]) > 0.8:  # Adjust cosine similarity threshold as needed
                identified_topics[topic] = {'count': count['count'] + 1, 'index': i}
                topic_found = True
                break
        if not topic_found:
            identified_topics[all_text[i]] = {'count': 1, 'index': i}

    # Extract identified topics and their respective phrases
    identified_topics_with_phrases = {}
    for topic, data in identified_topics.items():
        if data['count'] >= 2:  # Adjust threshold as needed
            identified_topics_with_phrases[topic] = {'count': data['count'], 'index': data['index']}

    return identified_topics_with_phrases

# Example usage
input_file = '/content/drive/MyDrive/Competitors Data.xlsx'  # Replace with your file path
df = detect_file_characteristics(input_file)

if df is not None:
    # Preprocess the data to fill NaN values
    df = df.fillna('Not found')

    # Process the data to identify topics dynamically
    identified_topics_with_phrases = process_data(df)

    # Convert the identified_topics dictionary to JSON format for easier reading
    identified_topics_complete = {topic: list(set(addresses)) for topic, addresses in identified_topics_with_phrases.items()}
    identified_topics_json = json.dumps(identified_topics_complete, indent=4)
    print(identified_topics_json)

    # Save to JSON file
    output_json_path = '/content/identified_topics.json'
    with open(output_json_path, 'w') as f:
        json.dump(identified_topics_complete, f, indent=4)
    print(f"Identified topics saved to: {output_json_path}")

    # Save to CSV file
    output_csv_path = '/content/identified_topics.csv'
    with open(output_csv_path, 'w', newline='') as csvfile:
        fieldnames = ['Topic', 'Count']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()

        for topic, data in identified_topics_with_phrases.items():
            writer.writerow({'Topic': topic, 'Count': data['count']})

    print(f"Identified topics saved to: {output_csv_path}")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.model_selection import GridSearchCV

data = pd.read_excel('/content/drive/MyDrive/Competitors Data.xlsx')

# Combine all text fields into a single column
data['combined_text'] = data['Address'].astype(str) + ' ' + data['Title 1'].astype(str) + ' ' + data['Meta Description 1'].astype(str) + ' ' + data['H1-1'].fillna('').astype(str)

# Create a document-term matrix
vectorizer = CountVectorizer(stop_words='english', max_df=0.95, min_df=2, max_features=1000)
doc_term_matrix = vectorizer.fit_transform(data['combined_text'])

# Define the range of n_components to test
param_grid = {'n_components': [10, 20, 30, 40, 50]}

# Initialize LDA model
lda = LatentDirichletAllocation(random_state=42)

# Perform grid search to find the best n_components
grid_search = GridSearchCV(lda, param_grid=param_grid, cv=5, verbose=1)
grid_search.fit(doc_term_matrix)

# Print the best n_components found
print("Best n_components:", grid_search.best_params_['n_components'])

# Fit LDA with the best n_components
best_lda_model = grid_search.best_estimator_
best_lda_output = best_lda_model.fit_transform(doc_term_matrix)

# Print the top words for each topic in the best model
feature_names = vectorizer.get_feature_names_out()
for topic_idx, topic in enumerate(best_lda_model.components_):
    top_words = [feature_names[i] for i in topic.argsort()[:-10 - 1:-1]]
    print(f"Topic {topic_idx + 1}: {', '.join(top_words)}")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.model_selection import GridSearchCV
from transformers import BartForConditionalGeneration, BartTokenizer
import torch
import numpy as np

# Load BART model and tokenizer
model_name = "facebook/bart-large-cnn"
tokenizer = BartTokenizer.from_pretrained(model_name)
model = BartForConditionalGeneration.from_pretrained(model_name)

# Function to generate topic name using BART
def generate_topic_name(topic, feature_names, model, tokenizer, num_words=30):
    # Get the indices of the top words sorted by importance
    top_word_indices = topic.argsort()[:-num_words-1:-1]

    # Get the words and their corresponding weights
    top_words = [feature_names[i] for i in top_word_indices]
    weights = topic[top_word_indices]

    # Normalize weights
    weights = weights / weights.sum()

    # Create a weighted string of words
    weighted_words = ' '.join([f"{word}:{weight:.2f}" for word, weight in zip(top_words, weights)])

    input_text = f"Summarize the following weighted keywords into a short, descriptive topic name. Keywords: {weighted_words}"
    inputs = tokenizer([input_text], max_length=1024, return_tensors="pt", truncation=True)
    summary_ids = model.generate(inputs["input_ids"], num_beams=4, min_length=5, max_length=20, early_stopping=True)
    topic_name = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return topic_name

# Load data from Excel
data = pd.read_excel('/content/drive/MyDrive/Competitors Data.xlsx')

# Combine relevant text fields into a single column
data['combined_text'] = data['Address'].astype(str) + ' ' + data['Title 1'].astype(str) + ' ' + data['Meta Description 1'].astype(str) + ' ' + data['H1-1'].fillna('').astype(str)

# Define additional stopwords for CountVectorizer
additional_stopwords = ['www', 'com', 'http', 'https']  # Add more as needed

# Create a document-term matrix
vectorizer = CountVectorizer(stop_words='english', max_df=0.95, min_df=2, max_features=1000,
                             token_pattern=r'\b[a-zA-Z]{3,}\b')  # Only words with 3 or more characters
doc_term_matrix = vectorizer.fit_transform(data['combined_text'])

# Define the range of n_components to test
param_grid = {'n_components': [10, 20, 30, 40, 50]}

# Initialize LDA model
lda = LatentDirichletAllocation(random_state=42)

# Perform grid search to find the best n_components
grid_search = GridSearchCV(lda, param_grid=param_grid, cv=5, verbose=1)
grid_search.fit(doc_term_matrix)

# Print the best n_components found
print("Best n_components:", grid_search.best_params_['n_components'])

# Fit LDA with the best n_components
best_lda_model = grid_search.best_estimator_
best_lda_output = best_lda_model.fit_transform(doc_term_matrix)

# Print the top words and assign semantically meaningful labels for each topic
feature_names = np.array(vectorizer.get_feature_names_out())
for topic_idx, topic in enumerate(best_lda_model.components_):
    # Generate a semantically meaningful topic name using BART
    topic_name = generate_topic_name(topic, feature_names, model, tokenizer)

    # Get top 10 words for display
    top_words_indices = topic.argsort()[:-11:-1]
    top_words = feature_names[top_words_indices]

    print(f"{topic_name}: {', '.join(top_words)}")
    print()
```

```python
import pandas as pd
import spacy
from collections import Counter
import re

# Load spaCy's English model
nlp = spacy.load("en_core_web_sm")

# Load the Excel file
data = pd.read_excel('/content/drive/MyDrive/Competitors Data.xlsx')

def clean_text(text):
    # Remove special characters and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text.lower().strip()

def extract_topics(text):
    # Process the text with spaCy
    doc = nlp(text)

    # Extract nouns and proper nouns
    topics = [token.lemma_ for token in doc if token.pos_ in ['NOUN', 'PROPN']]

    return topics

# Combine all text from Title, Meta Description, and H1
all_text = ""
for _, row in data.iterrows():
    all_text += clean_text(str(row.get('Title 1', ''))) + " "
    all_text += clean_text(str(row.get('Meta Description 1', ''))) + " "
    all_text += clean_text(str(row.get('H1-1', ''))) + " "

# Extract topics
topics = extract_topics(all_text)

# Count the most common topics
topic_counts = Counter(topics)
top_topics = topic_counts.most_common(20)  # Get top 20 topics

print("Top 20 topics for blog writing:")
for topic, count in top_topics:
    print(f"{topic}: {count}")

# Analyze semantic relationships
print("\nSemantic relationships:")
doc = nlp(all_text)
for sent in doc.sents:
    for token in sent:
        if token.dep_ in ['nsubj', 'dobj', 'pobj'] and token.head.pos_ == 'VERB':
            print(f"{token.text} ({token.dep_}) -> {token.head.text} (VERB)")

# Extract key phrases
print("\nKey phrases:")
key_phrases = []
for sent in doc.sents:
    for chunk in sent.noun_chunks:
        if len(chunk) > 1:  # Only phrases with more than one word
            key_phrases.append(chunk.text)

phrase_counts = Counter(key_phrases)
top_phrases = phrase_counts.most_common(10)
for phrase, count in top_phrases:
    print(f"{phrase}: {count}")
```
